Amb aquesta consideració es pot entendre que realitzar càlculs sobre fluxos de dades té la seva dificultat i pot ser impossible de trobar solucions exactes sense emmagatzemar tot el flux. Tot i això hi ha càlculs que sí es poden realitzar de forma exacta només considerant les dades que ens arriben. Un d'aquests càlculs és el de la mitjana aritmètica.
$$ \bar{x} = \frac{1}{W} \sum_{i=t-W+1}^{t}x_i$$

Observeu que en l'exemple anterior la mitjana en el temps t té un valor 0.39625 i la mitjana en el temps t+1 té un valor 0.31375.

Com podeu apreciar a la formula de la mitjana aritmètica, necessiteu guardar tot el flux de dades de la finestra t per el càlcul de la mitjana en el temps t+1. <b> Penseu e implementeu un mètode que ens permeti treure la mitjana en el temps t+1 a partir de la mitjana en el temps t.</b>


```python
class SlidWin:
    """
        Finestra lliscant que permet trobar la mitja d'una sèrie de dades de mida màxima W_size.
        Aquesta finestra lliscant recalcula la mitja amb les fòrmules d'increment i decrement que es troben més abaix d'aquest fitxer.
    """
    def __init__(self, W_size):
        """
            Constructor que inicialitza els paràmetres per la finestra lliscant.
            param W_size Mida màxima de la finestra lliscant.
        """
        self.data = [] # dades
        self.W_size = W_size # tamany de la finestra
        self.length = 0 # length actual de la finestra
        self.m = 0.0 # mitja de la finestra
    def add(self, x_t):
        """
            Afegeix una nova dada a la finestra i retorna la mitja d'aquesta i el valor de l'accio:.
            return 	new_mean,  1  : si la mitja ha disminuït
					new_mean, -1  : si la mitja ha augmentat
                    new_mean,  0  : si la mitja es identica
        """
        #Les primeres dades que encara no son de l'amplada de la finestra s'afegeixen de cop
        #i incrementen el valor del self.length
        if self.length < self.W_size:
            self.data.append(x_t)
            self.length += 1
        #A partir de que tenim la finestra "plena", cada vegada que entra una dada desplaçem les
        #que ja hi ha, eliminant-ne la mes vella i afegint-hi la mes nova
        else:
            for i in range(self.W_size-1):
                self.data[i] = self.data[i+1]
            self.data[self.length-1] = x_t
        
        #Calculem la mitja del vector
        new_mean = 0
        for i in range(self.length):
            new_mean += self.data[i]
        new_mean /= self.length
        
        old_mean = self.m
        #Assignem la nova mitja al valor de mitja de la classe
        self.m = new_mean
        
        if new_mean < old_mean:
            return new_mean, 1
        elif new_mean > old_mean:
            return new_mean, -1
        else:
            return new_mean, 0

        
#Testing de la funció
method1 = SlidWin(8)
v = [0.22, 0.88, 0.21, 0.41, 0.33, 0.41, 0.12, 0.43, 0.38, 0.22]

for i in range(len(v)):
    mitja, accio = method1.add(v[i])
    print "He afegit", v[i],"a la finestra a la iteració", i+1, "i m'ha retornat:"
    print " una mitja de", mitja, "i l'accio", accio
    print
```

```python
class AdWin:
	"""
		Finestra lliscant adaptativa que permet trobar la mitja de totes les dades que es van inserin i que permet trobar canvis bruscos
		en elles i adaptar-se a ells a partir d'un valor de confiança indicat per l'usuari.
	"""
	def __init__(self, d):
		"""
			Constructor de la finestra.
			
			param d Confiança el canvi. Com més gran més fàcil serà detectar un canvi.
		"""
		self.data = []
		self.length = 0
		self.rel = d
		self.m = 0.0
        ##
        self.W_list = list()
		
	def add(self, x_t):
		"""
			Afegeix una nova dada a la finestra i després de comprovar si hi ha hagut un canvi actualitza les dades necessàries.
			
			param x_t Nova dada.
			
			return self.m Mitja de les dades del interior de la finestra
			return 	0 si no hi ha hagut cap canvi significatiu
					1 si hi ha hagut un canvi i la mitja ha disminuït
					-1 si hi ha hagut un canvi i la mitja ha augmentat
		"""
        #Per defecte sempre afegirem un valor nou a la llista
        #i incrementarem la length(tamany actual de la finestra)
        """
        self.data.append(x_t)
        self.length += 1
        
        espera = True
        while espera:
            
		return self.m, 0
        
    def ecut(self, n1, n2):
        aux = 1 / 2 * calc_m(n1, n2)
        aux = aux * 4/delta_prime(n1+n2)
        return sqrt(aux)
    
    def calc_m(self, n1, n2):
        return 1/((1/n1) + (1/n2))
    
    def delta_prime(self, n):
        return sel.rel/n
        """
        m0 = self.m
        
        self.W_list,self.m = self.__ADWIN0(self.W_list,x_t,self.d,self.m)
        
        if (self.m-m0)>0:
            compra = 1
        else:
            compra = -1
            
        return self.m, compra

    
    def __ADWIN0(self, W_list, x_t, d, m):
        """
                It gets a new data from the adaptative sliding window and it tests for each possible division of it if there is a
                significative change of the average values m1 and m2, given the confidence parameter 'd'
                
                param W_list list with the windows data
                param x_t Float with the new data.
                param d confidence parameter.
                param m Average of the data W_list before the insertion.
                
                return W_list list with the data of the window updated
                return m Average of the list updated.
        """
        W_list.append(x_t)
        
        if len(W_list)<2:
            m = AverageIncremental(x_t,len(W_list)-1,m)
            return W_list,m

        m0 = None
        m1 = None
        
        for i in range(1,len(W_list)):
            W0 = W_list[:i]
            W1 = W_list[i:]
            if not m0: 
                m0 = sum(W0)/len(W0) #La primera vez que calculamos m0 y m1 lo haremos de 0, pero las siguientes veces podemos hacerlo de modo eficiente
            else: 
                m0 = AverageIncremental(W_list[i],i-1,m0)
            
            if not m1: 
                m1 = sum(W1)/len(W1)
            else: 
                m1 = AverageDecremental(W_list[i],len(W1)+1,m1)
                
            if math.fabs(m1-m0)>threshold(d,len(W0),len(W1)):
                m1 = AverageDecremental(W_list[i],len(W1)+1,m1)
                break
                        
        if W_list!=W1: 
            m = AverageIncremental(x_t,len(W_list)-1,m)  #Si no ha cortado la cola pre-W1 actualizamos la media de la lista con el nuevo valor
        
        if W_list!=W1: 
            m = AverageIncremental(x_t,len(W_list)-1,m)  #Si no ha cortado la cola pre-W1 actualizamos la media de la lista con el nuevo valor
        
        return W_list, m
        
        
def AverageIncremental(x_t, t, m):
        """
                Recompute the average 'm' when the data size increase of one element.
                
                param x_t Float, the new data 
                param t length of the data before inserting x_t
                param m average of the data before inserting x_t
                
                return float average updated 
        """
        return (m*t+x_t)/(t+1)
    
def AverageDecremental(x_t, t, m):
        """
                Recompute the average 'm' when the data size decreases of one element.
                
                param x_t Float, the new data 
                param t length of the data before deleting x_t
                param m average of the data before deleting x_t
                
                return float average updated 
        """
        return (m*t-x_t)/(t-1)

#Formula para el umbral de la ventana adaptativa
def threshold(d,n0,n1):
        m = 1.0/(1.0/n0+1.0/n1)
        delta = float(d)/(float(n0)+float(n1))
        E = math.sqrt(1.0/(2.0*m)*math.log(4.0/delta))
        return E
    
```
